**Data** **Preprocessing**

```python
items = df['0'].unique()
items
```

```python
encoded_vals = []
for index, row in df.iterrows():
    labels = {}
    uncommons = list(set(items) - set(row))
    commons = list(set(items).intersection(row))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
```

```python
ohe_df = pd.DataFrame(encoded_vals)
```

```python
ohe_df
```

**Association Rules**

```python
#For apriori, we need to do hot encoding.
freq_items = apriori(ohe_df, min_support = 0.2, use_colnames = True, verbose = 1)
```

```python
from itertools import combinations

rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)

# Create a DataFrame to store pairs of frequently associated items
associated_items = pd.DataFrame(columns=['Item1', 'Item2', 'Support', 'Confidence'])

# Iterate over each rule
for index, rule in rules.iterrows():
    antecedents = list(rule['antecedents'])
    consequents = list(rule['consequents'])

    # Combine antecedents and consequents to consider all items
    all_items = antecedents + consequents

    # Generate all possible pairs of items
    item_pairs = list(combinations(all_items, 2))

    # Add the pairs to the DataFrame
    for item1, item2 in item_pairs:
        support = rule['support']
        confidence = rule['confidence']

        # Add the pair to the DataFrame
        associated_items = associated_items.append({'Item1': item1, 'Item2': item2, 'Support': support, 'Confidence': confidence}, ignore_index=True)

```
